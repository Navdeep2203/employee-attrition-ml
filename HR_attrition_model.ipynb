```python
import pandas as pd

df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df['Attrition'].value_counts()
```

```python
df['Attrition'] = df['Attrition'].map({'No':0, 'Yes':1})
```

```python
df['Attrition'].head()
```

```python
df.drop(['EmployeeNumber','EmployeeCount','Over18','StandardHours'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.head()
```

```python
df.info()
```

```python
X = df.drop('Attrition' , axis = 1)
Y = df['Attrition']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
print(type(X))
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, Y,
    test_size=0.2,
    stratify=Y,
    random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)

lr.fit(X_train, y_train)

lr_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, lr_pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, lr_pred)

sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200, random_state=42)

rf.fit(X_train, y_train)

rf_pred = rf.predict(X_test)
```

```python
print(classification_report(y_test, rf_pred))
```

```python
cm = confusion_matrix(y_test, rf_pred)

sns.heatmap(cm, annot=True, fmt='d')
plt.show()
```

```python
rf = RandomForestClassifier(
    n_estimators=200,
    class_weight='balanced',
    random_state=42
)

rf.fit(X_train, y_train)

rf_pred = rf.predict(X_test)

print(classification_report(y_test, rf_pred))
```

```python
cm = confusion_matrix(y_test, rf_pred)

sns.heatmap(cm, annot=True, fmt='d')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score, roc_curve

probs = lr.predict_proba(X_test)[:,1]

auc = roc_auc_score(y_test, probs)
print("ROC-AUC:", auc)
```

```python
fpr, tpr, _ = roc_curve(y_test, probs)

plt.plot(fpr, tpr)
plt.plot([0,1],[0,1],'--')
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC Curve")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

importances = lr.coef_[0]   # since logistic is best
indices = np.argsort(np.abs(importances))[-10:]

plt.barh(range(len(indices)), importances[indices])
plt.yticks(range(len(indices)), [df.drop('Attrition',axis=1).columns[i] for i in indices])
plt.title("Top Features Influencing Attrition")
plt.show()
```
